How do computer see Images:
 1. They see it in two different way.

     a. **Grayscale** ü¶ú

     b. **Colored.** üçÄ

 2. In grayscale, Each Pixel represents the intensity of only one shade.
     that means how bright and dark the pixel is.

 3. Grayscale Images have One channel.

 4. Grayscale intensity range from (0, 255).

 5. Computers interpret grayscale images by analyzing these intensity values
      without considering color information.
---------------------------------------------------------------------------

 ‚óÄ**Colored Images**‚ñ∂  ‚õπ

1. Color images are more complex. They consist of **3 channel** Red, Blue and Green  ( Commonly known as RGB).

2. Each pixel in an RGB image has three intensity values, one for each channel.

3. Computer Processes the color of images by considering the combination of
  these 3 channel.

  ‚ö°**image read mode** ‚ö°

        using 1 shows the colored Image.
        using 0 shows the grayscale images.
        usign -1 shows the image including alpha channel.
  
* If the image has no alpha channel (e.g., JPEG), it reads the image as a 3-channel BGR image (Blue, Green, Red).

* If the image has an alpha channel (e.g., PNG or TIFF), it reads the image as a 4-channel BGRA image (Blue, Green, Red, Alpha).

* The **alpha channel** represents **transparency** in an image. ‚≠ï
